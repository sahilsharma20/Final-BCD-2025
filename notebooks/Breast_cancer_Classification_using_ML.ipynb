## Importing the Dependencies 

```python
import numpy as np
import pandas as pd

# importing scikit-learn 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

Data Collection and Processing

```python
# Loading the data to a data frame 

Breast_cancer_dataset = pd.read_csv("/Users/sahilsharma/Desktop/Final-BC/data/breast-cancer.csv.xls")
print ("Breast Cancer dataset is read successfully!")
print("\n")
# print(Breast_cancer_dataset)
Breast_cancer_dataset
```

```python
# ### Understanding the DataFrame Structure ###
# The `Breast_cancer_dataset` is already a Pandas DataFrame, so we directly assign it to `data_frame`.

# Check the type of the dataset to confirm it is a DataFrame
type(Breast_cancer_dataset)
```

```python

# Assign the dataset to a variable
data_frame = Breast_cancer_dataset  

# feature_names = Breast_cancer_dataset.columns.to_numpy()
# data_frame = pd.DataFrame(Breast_cancer_dataset.data, columns = Breast_cancer_dataset.feature_names)
# print(data_frame)
# print(feature_names)

# If you want to check column names (features)
data_frame.columns
```

```python
# print first 5 rows of the datafarme
data_frame.head()
```

```python
# adding the target column to the data frame have 
# data_frame["label"] = data_frame.target
```

```python
# Convert 'M' to 1 and 'B' to 0

label_encoder = LabelEncoder()
data_frame['diagnosis'] = label_encoder.fit_transform(data_frame['diagnosis'])

# If you want the last 5 rows
print("Last 5 Rows:\n")
data_frame.tail()
```

```python
# number of rows and columns in dataset
data_frame.shape
```


## <b>

<b>Dropping id column</b> <br><br>
Why is the 'id' column removed? 
- The 'id' column is just an identifier and does not provide meaningful information for predictions.
- Since it is not a feature that influences the classification task (Malignant/Benign), 
 <br> we **drop** it to ensure better model training and prevent unnecessary noise.


```python
# Drop 'id' column if present
data_frame = data_frame.drop(columns=['id'], errors='ignore')  
```

Separating the features and targets

```python
X = data_frame.drop(columns='diagnosis', axis=1)
Y = data_frame['diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

<b> Model Training </b>

<b>Logistic Regression</b>

```python
model = LogisticRegression()
```

```python
# training the Logistic regression using training data

model.fit(X_train, Y_train)
```

<b> Model Evaluation </b>

<b> Accuracy Score </b>

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print("Accuracy on Training data is equals to : " , (training_data_accuracy)*100)
```

```python
# accuracy on Testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print("Accuracy on Testing data is equals to : " , (testing_data_accuracy)*100)
```
